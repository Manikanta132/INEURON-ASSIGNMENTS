2. Write a Python program to reverse a word after accepting the input from the user.

```python
a = input("Please enter a word: ")
b = a[::-1]
print("The reverse word is: ",b)
```

1. Create the below pattern using nested for loop in Python. 
*
* *
* * *
* * * *
* * * * *
* * * *
* * *
* *
*

```python
a = range(1,6)

for i in a:
    print("* "*i)
    if i == 5:
        for j in a:
            print("* "*(i-j))
```
