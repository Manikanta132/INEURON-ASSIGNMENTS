### 1.2 Write a Python program to implement your own myfilter() function which works exactly like Python's built-in function filter()

```python
#Function for Filter function
def myfilter(func, seq):
    result = []
    for item in seq:
        if func(item):
            result.append(item)
    return result
```

```python
#Function to find a vowel in  a given list 
def vowel(l):
    for x in l:
        if x == 'a':
            return True
        elif x == 'e':
            return True
        elif x == "i":
            return True
        elif x == 'o':
            return True
        elif x == 'u':
            return True
        else:
            return False
```
